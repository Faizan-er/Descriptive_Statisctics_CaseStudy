# Boxplot

For some distributions/data sets, you will find that you need more information than the measures of central tendency (median, mean and mode). You need to have information on the variability or dispersion of the data. A boxplot is a graph that gives you a good indication of how the values in the data are spread out. Although boxplots may seem primitive in comparison to a histogram or density plot, they have the advantage of taking up less space, which is useful when comparing distributions between many groups or data sets.

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”).

# Theory:

##### Descriptive Statistics: Descriptive statistics involve the organization, summarization, and presentation of data in a meaningful way. It aims to describe the main features of a dataset without making any inferences or drawing conclusions beyond the data itself. Common types of descriptive statistics include:

- Measures of central tendency: These statistics describe the center or average of a dataset. They include the mean, median, and mode.
- Measures of variability: These statistics describe the spread or dispersion of a dataset. They include the range, variance, and standard deviation.
- Measures of shape or distribution: These statistics describe the shape of a dataset's distribution. They include skewness and kurtosis.
- Percentiles: These statistics divide a dataset into hundredths, allowing you to determine the position of a particular value within the dataset.
- Frequency distributions: These statistics organize data into groups or intervals and display the number of observations in each group.

##### Inferential Statistics: Inferential statistics involves drawing conclusions or making inferences about a population based on a sample of data. It allows researchers to generalize their findings beyond the observed data and make predictions or test hypotheses. Some common types of inferential statistics include:

- Confidence intervals: These statistics estimate a range of values within which a population parameter is likely to fall.
- Hypothesis testing: This involves testing a research hypothesis by comparing sample data to a known or assumed population parameter.
- Regression analysis: Regression allows you to explore relationships between variables and make predictions about one variable based on another.
- Analysis of variance (ANOVA): ANOVA is used to test the differences between means of two or more groups.
- Chi-square tests: These tests examine the association or independence between categorical variables.

These are just a few examples of the types of statistics within each category. Statistics is a broad field with various techniques and methods that are used for different purposes depending on the nature of the data and the research questions being addressed.

## Predictive and prescriptive statistics fall under the category of inferential statistics.

- Predictive Statistics: Predictive statistics involve using historical or existing data to make predictions or forecasts about future events or outcomes. It aims to estimate or predict unknown or future values based on patterns observed in the data. Predictive statistics often use techniques such as regression analysis, time series analysis, and machine learning algorithms to create models that can forecast future trends or make predictions about specific variables.

- Prescriptive Statistics: Prescriptive statistics goes beyond prediction and involves providing recommendations or guidance on the best course of action based on the available data. It helps in decision-making by considering various possible actions and their potential outcomes. Prescriptive statistics often involve optimization techniques, simulation methods, and decision analysis to identify the best choices or strategies to achieve desired goals.

###### Both predictive and prescriptive statistics rely on inferential techniques to draw conclusions or make predictions based on sample data. They aim to go beyond descriptive statistics by providing insights into future trends, behavior, or decision-making, allowing individuals or organizations to make informed choices or take appropriate actions.
